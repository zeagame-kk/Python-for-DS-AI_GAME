### Perform the SVM classification using our scratch code; put it into class, and allows users to use soft or hard margin

```python
import sys
sys.path.append('C:\Python27\Lib\site-packages')

import cvxopt
from numpy import linalg
from cvxopt import solvers as cvxopt_solvers

class SVMModel:
   
    def linear(self, x, z):
        return np.dot(x, z.T)

    def polynomial(self, x, z, p=2):
        return (1 + np.dot(x, z.T)) ** p

    def gaussian(self, x, z, sigma=0.1):
        return np.exp(-linalg.norm(x - z) ** 2 / (2 * (sigma ** 2)))

    def fit(self, X, y, kernel, C):
        self.y = y
        self.X = X
        n_samples, n_features = X.shape

        # Calculate Kernel
        K = np.zeros((n_samples, n_samples))
        for i in range(n_samples):
            for j in range(n_samples):
                if kernel == 'linear':
                    K[i,j] = self.linear(X[i], X[j])
                elif kernel == 'polynomial':
                    K[i,j] = self.polynomial(X[i], X[j])
                else:
                    K[i,j] = self.gaussian(X[i], X[j])
            
        # Solve with cvxopt final QP needs to be reformulated
        # to match the input form for cvxopt.solvers.qp
        P = cvxopt.matrix(np.outer(y, y) * K)
        q = cvxopt.matrix(-np.ones((n_samples, 1)))
#         G = cvxopt.matrix(np.vstack((np.eye(n_samples) * -1, np.eye(n_samples))))
#         h = cvxopt.matrix(np.hstack((np.zeros(n_samples), np.ones(n_samples) * self.C)))
        A = cvxopt.matrix(y, (1, n_samples), "d")
        b = cvxopt.matrix(np.zeros(1))
        
        #from solutions---------------------------------------------
        if C is None:
            G = cvxopt.matrix(np.diag(np.ones(n_samples) * -1))
            h = cvxopt.matrix(np.zeros(n_samples))
        else:
            tmp1 = np.diag(np.ones(n_samples) * -1)
            tmp2 = np.identity(n_samples)
            G = cvxopt.matrix(np.vstack((tmp1, tmp2)))
            tmp1 = np.zeros(n_samples)
            tmp2 = np.ones(n_samples) * C
            h = cvxopt.matrix(np.hstack((tmp1, tmp2)))
        #------------------------------------------------------------
        
        cvxopt.solvers.options["show_progress"] = False
        
        sol = cvxopt.solvers.qp(P, q, G, h, A, b)
        
        a = np.ravel(sol["x"])
        
        # Support vectors have non zero lagrange multipliers
        sv_idx = a > 1e-5
        ind = np.arange(len(a))[sv_idx]
        a = a[sv_idx]    
        sv = X[sv_idx]
        sv_y = y[sv_idx]
        print("%d support vectors out of %d points" % (len(a), n_samples))

        # Intercept
        b = 0
        for n in range(len(a)): #b = y - wTx
            b += sv_y[n]
            b -= np.sum(a * sv_y * K[ind[n],sv_idx])
        b /= len(a)

        # Weight vector
        if kernel == 'linear':
            w = np.zeros(n_features)
            for n in range(len(a)):
                w += a[n] * sv_y[n] * sv[n]
        else:
            w = None
        return sv, sv_y, a, w, b
    
    #-------------------------------------------------------------------------------
    def project(self,X, kernel,sv, sv_y, a, w, b):
        #kernel = linear
        if w is not None:  #this is linear!!!!!
            return np.dot(X, w) + b
        else:
            y_predict = np.zeros(len(X))
            for i in range(len(X)):
                s = 0
                #The zip() function returns a zip object, which is an iterator of tuples
                for a_val, sv_y_val, sv_val in zip(a, sv_y, sv):   
                    if kernel == 'polynomial':
                        s += a_val * sv_y_val * self.polynomial(X[i], sv_val)
                    else:
                        s += a_val * sv_y_val * self.gaussian(X[i], sv_val)
                y_predict[i] = s
            return y_predict + b

    def predict(self,X, kernel, sv, sv_y, a, w, b):
        return np.sign(self.project(X, kernel,sv, sv_y, a, w, b))
    
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset_np[:,[0,1]], dataset_np[:,2], test_size=0.20, random_state=42)

y_train[y_train==0] = -1
y_test[y_test==0] = -1
```

```python
SVM = SVMModel()

kernel = 'gaussian'

sv, sv_y, a, w, b = SVM.fit(X_train, y_train, kernel=kernel, C=None)
```

```python
y_pred = SVM.predict(X_test, kernel, sv, sv_y, a, w, b).astype(int)

print("y_test:",y_test.astype(int))
print("y_pred:",y_pred)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
